```python
# from textblob import TextBlob
import pandas as pd

# from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from transformers import pipeline

# Load the CSV file into a DataFrame
df = pd.read_csv('final_data.csv')


# # Initialize the VADER sentiment analyzer
# analyzer = SentimentIntensityAnalyzer()

# def get_sentiment(text):
#     score = analyzer.polarity_scores(text)
#     if score['compound'] > 0.05:
#         return 1  # Positive sentiment
#     elif score['compound'] < -0.05:
#         return -1  # Negative sentiment
#     else:
#         return 0  # Neutral sentiment



# # Initialize a sentiment-analysis pipeline
# sentiment_analyzer = pipeline('sentiment-analysis')

# def get_sentiment(text):
#     result = sentiment_analyzer(text)[0]
#     if result['label'] == 'POSITIVE':
#         return 1  # Positive sentiment
#     elif result['label'] == 'NEGATIVE':
#         return -1  # Negative sentiment
#     else:
#         return 0  # Neutral sentiment

import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.nn.functional import softmax

# Load the pre-trained BERT model and tokenizer
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"  # A model fine-tuned for sentiment analysis
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Define the sentiment analysis function
def get_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    scores = softmax(outputs.logits, dim=1).numpy()[0]

    # Determine sentiment based on the scores
    if scores[4] > 0.5:  # Positive
        return 1
    elif scores[2] > 0.5:  # Neutral
        return 0
    else:  # Negative
        return -1



df['sentiment'] = df['TweetText'].apply(get_sentiment)
df['Jio_label'] = df.apply(lambda row: row['sentiment'] if 'jio' in row['TweetText'].lower() else '',axis=1)
df['Airtel_label'] = df.apply(lambda row: row['sentiment'] if 'airtel' in row['TweetText'].lower() else '', axis=1)
df['Vodafone_label'] = df.apply(lambda row: row['sentiment'] if 'vodafone' in row['TweetText'].lower() else '', axis=1)


df.to_csv('final1_data.csv',index=False)
```
